## explicacion dtype e indexacion

* **dtype:** la librería pandas (pd) infiere automáticamente los tipos de datos. Esto se hace para optimizar la memoria y el rendimiento. Por ejemplo, `int64` corresponde a enteros y `float` a números decimales.

#### diferencia entre DataFrame y Series

como vimos en clases, se explicó el significado de PanDaS:

* Pan: hace referencia a panel  
* Da: nos habla de data o datos  
* S: hace referencia a series  

**Series:** una serie es un array de una dimensión y es capaz de contener cualquier tipo de dato. En Excel sería como una columna que almacena datos. Cada elemento de una serie tiene un "índice", conocido como etiqueta.

**DataFrame:** es una estructura de datos de dos dimensiones etiquetada con columnas. Podemos entenderlo como una tabla SQL o una hoja de Excel. Es una colección de series que comparten el mismo índice.  **(Un ejemplo sería el que construimos anteriormente)**

## Cómo influye el tipo de dato en la eficiencia computacional

El tipo de dato `dtype` tiene varios impactos en la eficiencia computacional:

**Uso de memoria:**

* **Tipos de datos más pequeños:** existen diferentes tipos como `int8`, `int32`, `int64`. Mientras más pequeño el tipo, menor uso de memoria RAM, lo que puede hacer que las operaciones sean más rápidas.

* **Category dtype:** para columnas con un número limitado de valores únicos, convertir el tipo de dato a `category` puede reducir bastante el uso de memoria y acelerar operaciones como `groupby`.

**Velocidad de operaciones:**

* **Operaciones numéricas:** las operaciones matemáticas en columnas numéricas (`int`, `float`) son mucho más rápidas que en columnas de tipo `object` (texto). Esto se debe a que pandas y NumPy están optimizados para trabajar con datos numéricos.

* **Comparaciones y filtrado:** son más eficientes con datos numéricos o booleanos que con texto, especialmente si las cadenas son largas.

* **Missing values:** el manejo de valores nulos también influye. Por ejemplo, en datos numéricos se usa `NaN`, mientras que en `object` puede usarse `None` o texto vacío, lo que puede afectar el rendimiento.

**Compatibilidad y errores:**

* **Operaciones inválidas:** realizar operaciones numéricas en columnas con tipos incorrectos puede generar errores y obliga a convertir datos, lo que consume tiempo.

* **Consistencia:** mantener tipos de datos consistentes ayuda a evitar errores y asegura que las funciones de `pandas` funcionen correctamente.

Dado un DataFrame de ventas con columnas: Cliente, Producto, Monto, Fecha: (cree uno con al menos 20 registros)

### Tarea:

*   Seleccione todos los registros donde:
    *   Monto > 50000 y Producto == "Laptop"
*   Modifique el valor del Monto en un 10% de aumento para dichos registros
*   Explique por qué `loc` es preferible frente a indexación tradicional en este caso

##solucion
### creacion del Df de ventas

importar librerias `pandas` y `numpy`


```python
import pandas as pd
import numpy as np

#se crean datos aleatorios
np.random.seed(42)

#usamos 25 registros para continuar
n_registros = 25
#usamos un for para ir por cada cliente
clientes = [f'Cliente_{i}' for i in range(1, n_registros + 1)]
#asignamos los productos mediante una lista
productos_lista = ['Laptop', 'Teclado', 'Mouse', 'Monitor', 'Webcam']

# Crea un rango de fechas con las siguientes características:
# start: Comienza el 1 de enero de 2023
# periods: Genera exactamente la cantidad de fechas que indica la variable 'n_registros'
# freq='D': El salto entre cada fecha es de un día (Diario)
fechas = pd.to_datetime(pd.date_range(start='2023-01-01', periods=n_registros, freq='D'))

#aignamos datos a cada fila con su respectiva etiqueta
df_ventas = pd.DataFrame({
    'Cliente': np.random.choice(clientes, size=n_registros),
    'Producto': np.random.choice(productos_lista, size=n_registros),
    'Monto': np.random.randint(10000, 100000, size=n_registros),
    'Fecha': fechas
})
# hacemos que los apartados de laptop esten fijo (para tener dos >50.000)
df_ventas.loc[5, ['Producto', 'Monto']] = ['Laptop', 60000]
df_ventas.loc[10, ['Producto', 'Monto']] = ['Laptop', 75000]

#imprime el df por pantalla
print("DataFrame de Ventas (Primeros 12 registros):\n")
print(df_ventas.head(12))
```

### Se seleccionan los registros donde el `Monto` sea mayor a 50000 y el `Producto` sea "Laptop". Luego, aumentaremos el `Monto` de estos registros en un 10%.

```python
# Definir la condición
condicion = (df_ventas['Monto'] > 50000) & (df_ventas['Producto'] == 'Laptop')

# Guardamos una copia ANTES de modificar (importante para comparar)
df_filtrado_antes = df_ventas[condicion].copy()

print("\nRegistros que cumplen la condición ANTES del aumento:\n")
print(df_filtrado_antes)

# Aplicar el aumento del 10% usando .loc
# .loc[filas, columna] garantiza que modificamos el DataFrame original
df_ventas.loc[condicion, 'Monto'] *= 1.10

print("\nRegistros que cumplen la condición DESPUÉS del aumento (10% extra):\n")
print(df_ventas.loc[condicion])
```

####explicación: ¿Por qué loc es preferible frente a la indexación tradicional?

El uso de `.loc` es la forma estándar y recomendada en pandas por las siguientes razones:

* **Evita el error de copia:** la indexación tradicional como `df[condicion]['Monto']` realiza lo que se conoce como "indexación encadenada". En este caso, pandas no puede asegurar si se está trabajando sobre una copia o una vista del DataFrame original, lo que puede provocar que los cambios no se guarden correctamente.

* **Claridad:** con `.loc[filas, columnas]` se indica explícitamente qué filas y qué columna se desean modificar dentro del DataFrame original.

* **Consistencia:** `.loc` es más predecible, ya que siempre modifica directamente el DataFrame, evitando comportamientos inesperados.

## diferencia entre loc e iloc
Como revisamos antes en DataFrames, existen dos formas para extraer datos:

*  **loc (Localización por etiqueta):** Busca la información utilizando los nombres explícitos (etiquetas) asignados a las filas o columnas. Una característica clave es que al usar rangos (ej: `loc['A':'C']`), el último elemento del rango **sí** se incluye en el resultado final.

*  **iloc (Localización posicional):** Busca la información basándose estrictamente en su posición numérica entera (0, 1, 2...), tal como funcionan las listas tradicionales. A diferencia de `loc`, al usar rangos (ej: `iloc[0:3]`), el límite superior **no** se incluye (solo traería las posiciones 0, 1 y 2).

## Riesgos de errores silenciosos en indexación posicional

El uso de (`iloc`) tiene ciertos riesgos si no tenemos control absoluto sobre la estructura de los datos :

**¿Qué es un error silencioso?**
*  **Concepto:** Ocurre cuando un bloque de código se ejecuta "exitosamente" sin lanzar ninguna alerta, detención o mensaje de error en pantalla, pero el resultado de los datos es incorrecto.

**El problema al depender de iloc:**
*  **Sensibilidad a cambios de estructura:** `iloc` asume que el orden del dataset es totalmente estático. Si el dataset cambia su estructura (por ejemplo, si se agrega una columna nueva al principio o se reordenan), el código seguirá funcionando pero **seleccionará datos incorrectos**. Esto se llama error silencioso porque no arroja un mensaje de error, pero los resultados del análisis serán falsos.

##pregunta 4: análisis de Comportamiento de Clientes

**Realice:**

* Segmentación por percentiles (ej: top 25% de ingresos)
* Explique cómo el slicing permite construir variables derivadas útiles para modelos predictivos


Se analiza un dataset de comportamiento de clientes para identificar segmentos específicos mediante filtrado y uso de percentiles.

---


```python
import pandas as pd
import numpy as np

#usamos la url que es de donde obtenemos los datos para seguir con el analisis
url = "https://raw.githubusercontent.com/tirthajyoti/Machine-Learning-with-Python/master/Datasets/Mall_Customers.csv"
df = pd.read_csv(url )
regiones = ['Norte', 'Sur', 'Este', 'Oeste']
#datos aleatorios
np.random.seed(42)
# punto importante es añadir una columna region, ya que no existe
df['Region'] = np.random.choice(regiones, size=len(df))

#filtrado multiple bajo ciertas condiciones
condicion = (df['Age'] > 30) & (df['Region'] == 'Norte') & (df['Annual Income (k$)'] > 50)
#se aplica la condicion
clientes_filtrados = df.loc[condicion]

#segmentacion por percentiles
umbral = df['Annual Income (k$)'].quantile(0.75)
#se filtra el data set original para que cumplan la condicion
top_clientes = df.loc[df['Annual Income (k$)'] >= umbral]

#imprimir los primeros 5 resultados
print("Clientes Filtrados (Primeros 5):\n")
print(clientes_filtrados.head())
print(f"\nClientes Top 25% Ingresos (Umbral: {umbral}k$):\n")
print(top_clientes.head())
#conteo clientes
print("\nCantidad de clientes filtrados:", len(clientes_filtrados))
print("Cantidad de clientes top 25%:", len(top_clientes))

```

##Explicación: Slicing y Variables Derivadas

El **slicing** (o recorte de datos) es fundamental para construir variables derivadas útiles en modelos predictivos por las siguientes razones:

1.  **Creación de Etiquetas de Segmento**: Permite dividir una variable continua (como el ingreso) en categorías (ej: "Ingreso Alto", "Ingreso Medio", "Ingreso Bajo"). Estas etiquetas ayudan a los modelos a capturar patrones no lineales.
2.  **Aislamiento de Comportamientos**: Al recortar datos por periodos de tiempo o regiones, podemos crear variables como "Gasto promedio en los últimos 3 meses", lo cual es un predictor mucho más potente que el gasto total histórico.
3.  **Reducción de Ruido**: Permite eliminar valores atípicos (outliers) mediante el recorte de percentiles extremos, mejorando la precisión y estabilidad de los modelos de Machine Learning.

##pregunta 5: medidas de tendencia central

Calcule:
* Media, mediana y moda
* Compare:
Resultados usando pandas vs cálculo manual

* Interprete:
  Qué ocurre en presencia de outliers

  Cuándo es preferible cada medida

se analiza el dataset Titanic para comparar los métodos de cálculo y entender el impacto de los valores extremos.

---



```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url )
data = df['Fare'].dropna() # seleccionar columna de tarifa para trabjar en ella

print(data.head(5)) # lo suamos para visualizar los datos

# usando la libreria pandas(pd) es relativamente sencillo calcular las 3
media_pd = data.mean()
mediana_pd = data.median()
# usamos [0] porque mode() puede devolver más de un valor
moda_pd = data.mode()[0]

# calculo manual usando fuunciones y logica:
# media
media_manual = sum(data) / len(data)

# mediana
data_sorted = sorted(data)
n = len(data_sorted)
if n % 2 == 0:
    mediana_manual = (data_sorted[n//2 - 1] + data_sorted[n//2]) / 2
else:
    mediana_manual = data_sorted[n//2]

# calculo de la moda
frecuencias = {}
for valor in data:
    frecuencias[valor] = frecuencias.get(valor, 0) + 1

# buscamos la frecuencia con el valor maximo
moda_manual = max(frecuencias, key=frecuencias.get)

print("\ncomparación de resultados")
print(f"Media   = Pandas: {media_pd:.2f} | Manual: {media_manual:.2f}")
print(f"Mediana = Pandas: {mediana_pd:.2f} | Manual: {mediana_manual:.2f}")
print(f"Moda    = Pandas: {moda_pd:.2f}  | Manual: {moda_manual:.2f}")

print("\nResumen estadístico:\n")
print(data.describe())
```

#### Comparación uso de: pandas vs manual

*  **Resultados:** Ambos métodos arrojan exactamente los mismos valores, lo que valida la lógica matemática detrás de las funciones de Pandas.
*  **Eficiencia:** El cálculo manual requiere iteraciones explícitas (bucles `for`, ordenamiento manual), lo cual es ineficiente en grandes volúmenes de datos. Pandas utiliza operaciones vectorizadas, haciendo que el cálculo sea casi instantáneo con menos líneas de código.

## Interpretación de Medidas de Tendencia Central



### ¿Qué ocurre en presencia de outliers?
En este dataset, el valor máximo es **512.33**, mientras que la mediana es solo **14.45**.

* La **media** se ve "arrastrada" hacia arriba por los valores altos (outliers), dando una impresión de que el pasaje promedio era más caro de lo que realmente era.
* La **mediana** se mantiene estable, mostrando que el 50% de los pasajeros pagó menos de 14.45, lo cual representa mejor la realidad del conjunto de datos.

### ¿Cuándo es preferible cada medida?
1.  **Media**: Es preferible cuando los datos son **simétricos** y no tienen valores extremos (ej: alturas de personas).
2.  **Mediana**: Es la mejor opción cuando hay **outliers** o datos sesgados (ej: ingresos, precios de viviendas), ya que no se ve afectada por valores inusualmente altos o bajos.
3.  **Moda**: Es útil para variables **categóricas** o cuando queremos saber cuál es el valor más común o frecuente en una serie (ej: la talla de zapato más vendida).

## pregunta 6: Tabla y frecuencias

Construya:
* Tabla de frecuencia absoluta
* Frecuencia relativa
* Frecuencia acumulada
* Analice:
 Cómo estas tablas pueden apoyar decisiones de negocio

la construcción y análisis de tablas de frecuencia de una variable

---

```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# se usa clase de billete para analisis
frec_abs = df['Pclass'].value_counts().sort_index()
frec_rel = df['Pclass'].value_counts(normalize=True).sort_index()
frec_acum = frec_abs.cumsum()

# se construye la tabla
tabla_frecuencias = pd.DataFrame({
    'Frecuencia Absoluta': frec_abs,
    'Frecuencia Relativa (%)': (frec_rel * 100).round(2),
    'Frecuencia Acumulada': frec_acum
})

tabla_frecuencias.index.name = 'Clase de Pasajero'

# esto es algo que vimos en una clase cuando se mostro el histograma o en frecuencias
# no alcanzaba el display y se edito de una manera que se pudiese visualizar completo
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("tabla de frecuencias\n")
print(tabla_frecuencias)
```

## analisis: uso de tablas de frecuencia en decisiones de negocio

las tablas de frecuencia permiten entender rapidamente como se distribuyen los datos dentro de una variable, en este caso la clase de pasajero (`Pclass`).

* **frecuencia absoluta:** muestra cuántos pasajeros hay en cada clase. esto permite identificar qué segmento es más común dentro del dataset.

* **frecuencia relativa:** indica el porcentaje que representa cada clase respecto al total. esto es útil para comparar proporciones y entender el peso real de cada grupo.

* **frecuencia acumulada:** permite ver cómo se van sumando los datos hasta cierto punto. aunque en variables categóricas no siempre es tan relevante, puede ayudar a ver concentraciones.

### aplicacion en decisiones de negocio

* **segmentacion de clientes:** permite identificar cuál es el grupo predominante (por ejemplo, más pasajeros en clase baja), lo que ayuda a enfocar estrategias en ese segmento.

* **definicion de precios:** si la mayoría pertenece a clases más económicas, se pueden ajustar precios o promociones para ese grupo.

* **optimizacion de recursos:** entender la distribución ayuda a asignar mejor recursos (espacio, servicios, atención) según la demanda real.

* **deteccion de oportunidades:** si un segmento tiene baja representación, puede ser una oportunidad de crecimiento o expansión.

en resumen, estas tablas permiten transformar datos en informacion util para tomar decisiones más informadas y reducir la incertidumbre.

```python
import pandas as pd
import numpy as np

# usamos un dataset Titanic adaptado a un contexto de Marketing
# usamos un url para tener el df
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
# usamos df.read para poder leer el archivo.csv
df = pd.read_csv(url )

#pritn para visualizar la carga
print("Dataset cargado exitosamente\n")
print(f"Total de registros: {len(df)}")

# Eliminamos filas con valores nulos en columnas
df = df.dropna(subset=['Age', 'Fare', 'Embarked'])

# Cambiamos el nombre a minusculas para facilitar la busqueda
df.columns = [col.strip().lower() for col in df.columns]

# filtrado con .iloc y .loc
# Uso de loc: Clientes mayores de 30 años que pagaron una tarifa alta (Segmento premium)
filtro_loc = df.loc[(df['age'] > 30) & (df['fare'] > 50)]

# Uso de iloc: Seleccionar las primeras 5 filas y las primeras 4 columnas para inspección
filtro_iloc = df.iloc[0:5, 0:4]

#print para visualizar el filtrado
print("\nResultado filtrado loc (Segmento premium)\n")
print(filtro_loc[['age', 'fare', 'sex']].head())

# caluclos estadisticos
# Obtenemos promedios, valores mínimos y máximos de edad y gasto (Fare)
estadisticos = df[['age', 'fare']].describe()

#print para ver los resultados del analisis
print("\nEstadísticos descriptivos (Edad y gasto)\n")
print(estadisticos)

# construir tablas de frecuencia
# Tabla de frecuencia absoluta para la variable 'pclass' (Clase/nivel socioeconómico)
frec_absoluta = df['pclass'].value_counts().sort_index()

# Tabla de frecuencia relativa (%)
frec_relativa = (df['pclass'].value_counts(normalize=True) * 100).sort_index().round(2)

#tabla asociada a sus respectivos nombres
tabla_frecuencia = pd.DataFrame({
    'Frecuencia absoluta': frec_absoluta,
    'Frecuencia relativa (%)': frec_relativa
})
#cambio de index por nivel clase
tabla_frecuencia.index.name = 'Nivel/Clase'

print("\nTabla de Frecuencia: Niveles Socioeconómicos\n")
print(tabla_frecuencia)

# GENERACIÓN DE INSIGHTS
# Insight 1: Gasto promedio por género
gasto_genero = df.groupby('sex')['fare'].mean()

# Insight 2: Tasa de conversión (survived) general
tasa_conversion = df['survived'].mean() * 100

print("\nINSIGHTS ESTRATÉGICOS\n")
print(f"1. El gasto promedio (Fare) de las mujeres es {gasto_genero['female']:.2f} vs {gasto_genero['male']:.2f} de los hombres.")
print(f"2. La tasa de conversión general de la campaña es del {tasa_conversion:.2f}%.")
```
